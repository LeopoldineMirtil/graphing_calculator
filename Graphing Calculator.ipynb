# Graphing Calculator

```python
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, linsolve
import math

def graph_calc():

    print("Graphing Calculator Menu: \n",
          "1. Table of Values                               6. Solve System of Equations\n",
          "2. Graph and Table of Values                     7. Shade Above/Below Line \n",
          "3. Plot Equations and Intersection Point         8. Solve and Graph System of Equations\n",
          "4. Plot Roots and Vertex of Quadratic Function   9. Zoom In/Out on Graph\n",
          "5. Plot Functions                               10. Solve Quadratic Equation\n")

    while True:
        #user selection
        user_choice = int(input("Enter the number(1-10) of your choice from the menu: "))

        if user_choice==1:
            def table():
                #get function input
                mb = input("Enter m and b, separated by a comma, to complete y = mx + b equation: ")
                mb_split = mb.split(",")
                m = float(mb_split[0])
                b = float(mb_split[1])

                ax = plt.subplot()
                ax.set_axis_off()

                if b < 0:
                    minus_split = str(b).split("-")
                    neg_b = float(minus_split[1])
                    title = f'y = {int(m)}x - {int(neg_b)}'
                else:
                    title = f'y = {int(m)}x + {int(b)}'

                cols = ('x', 'y')
                rows = [[0,0]]
                for a in range(1,10):
                    rows.append([a, int(m)*a + int(b)]) 

                ax.set_title(title)
                plt.table(cellText=rows, colLabels=cols, cellLoc='center', loc='upper left')
                plt.show() 
            table()
            break

        elif user_choice==2:
            def graph_table():
                mb = input("Enter m and b, separated by a comma, to complete y = mx + b equation: ")
                mb_split = mb.split(",")
                m = float(mb_split[0])
                b = float(mb_split[1])

                #format table title
                if b < 0:
                    minus_split = str(b).split("-")
                    neg_b = float(minus_split[1])
                    title = f'y = {int(m)}x - {int(neg_b)}'
                else:
                    title = f'y = {int(m)}x + {int(b)}'

                #line graph
                xmin = -20
                xmax = 20
                ymin = -20
                ymax = 20

                points = 20*(xmax-xmin)
                x = np.linspace(xmin,xmax,points)

                fig, ax = plt.subplots()
                plt.axis([xmin,xmax,ymin,ymax]) 
                plt.plot([xmin,xmax],[0,0],'k') 
                plt.plot([0,0],[ymin,ymax], 'k') 
                plt.plot(x, m*x + b)
                ax.set_title(title)
                plt.show()

                #table of values
                ax = plt.subplot()
                ax.set_axis_off()

                cols = ('x', 'y')
                rows = [[0,0]]
                for a in range(1,10):
                    rows.append([a, int(m)*a + int(b)]) 

                ax.set_title(title)
                plt.table(cellText=rows, colLabels=cols, cellLoc='center', loc='upper left')
                plt.show() 
                            
            graph_table()
            break

        elif user_choice==3:
            def intersect():

                #1st eq input
                mb_1 = input("First equation: y = mx + b. Enter m and b, separated by a comma: ")
                mb_in1 = mb_1.split(",")
                m1 = float(mb_in1[0])
                b1 = float(mb_in1[1])

                #2nd eq input
                mb_2 = input("Second equation: y = mx + b. Enter m and b, separated by a comma: ")
                mb_in2 = mb_2.split(",")
                m2 = float(mb_in2[0])
                b2 = float(mb_in2[1])


                #print out equations
                if b1 < 0 and b2 > 0:
                    minus_split = str(b1).split("-")
                    neg_b1 = float(minus_split[1])
                    print(f'First equation:   y = {int(m1)}x - {int(neg_b1)}')
                    print(f'Second equation:  y = {int(m2)}x + {int(b2)}')
                
                elif b1 > 0 and b2 < 0:
                    minus_split = str(b2).split("-")
                    neg_b2 = float(minus_split[1])
                    print(f'First equation:   y = {int(m1)}x + {int(b1)}')
                    print(f'Second equation:  y = {int(m2)}x - {int(neg_b2)}')

                elif b1 < 0 and b2 < 0:
                    minus_split = str(b1).split("-")
                    neg_b1 = float(minus_split[1])
                    minus_split = str(b2).split("-")
                    neg_b2 = float(minus_split[1])
                    print(f'First equation:   y = {int(m1)}x - {int(neg_b1)}')
                    print(f'Second equation:  y = {int(m2)}x - {int(neg_b2)}')

                else:
                    print(f'First equation:   y = {int(m1)}x + {int(b1)}')
                    print(f'Second equation:  y = {int(m2)}x + {int(b2)}')


                # solve system of eqs
                x,y = symbols('x y')
                first = m1*x + b1 - y
                second = m2*x + b2 - y
                solution = linsolve([first, second], (x, y))
                x_solution = round(float(solution.args[0][0]),3)
                y_solution = round(float(solution.args[0][1]),3)

                #ensure window includes solution
                xmin = int(x_solution) - 10
                xmax = int(x_solution) + 10
                ymin = int(y_solution) - 10
                ymax = int(y_solution) + 10
                points = 2*(xmax-xmin)

                graph_x = np.linspace(xmin,xmax,points)

                #y values for the graph
                y1 = m1*graph_x + b1
                y2 = m2*graph_x + b2

                fig, ax = plt.subplots()
                plt.axis([xmin,xmax,ymin,ymax]) 
                plt.plot([xmin,xmax],[0,0],'k') 
                plt.plot([0,0],[ymin,ymax], 'k') 

                # plot line 1
                plt.plot(graph_x, y1, 'g')

                # plot line 2
                plt.plot(graph_x, y2, 'b')

                # intersection point
                plt.plot([x_solution],[y_solution],'ro')

                plt.show()
            intersect()
            break

        elif user_choice==4:
            def quad():
                #user input for quad eq
                a = int(input("For y = ax\u00b2 + bx + c, enter the 'a' value: "))
                b = int(input("For y = ax\u00b2 + bx + c, enter the 'b' value: "))
                c = int(input("For y = ax\u00b2 + bx + c, enter the 'c' value: "))

                print(f'Quadratic equation: y = {a}x\u00b2 + {b}x + {c}')

                # vertex points
                vx = round(-b/(2*a), 2)
                vy = round(a*vx**2 + b*vx + c, 2)
                print(f'Vertex: ({int(vx)}, {int(vy)})')

                xmin = int(vx)-20
                xmax = int(vx)+20
                ymin = int(vy)-20
                ymax = int(vy)+20
                points = 20*(xmax-xmin)
                x = np.linspace(xmin,xmax,points)

            
                fig, ax = plt.subplots()
                plt.axis([xmin,xmax,ymin,ymax]) 
                plt.plot([xmin,xmax],[0,0],'k') 
                plt.plot([0,0],[ymin,ymax], 'k') 

                # plot vertex
                plt.plot([vx],[vy],'ro') 

                #plot parabola
                y = a*x**2 + b*x + c
                plt.plot(x,y)

                #find and plot the roots
                d = b**2 - 4*a*c
                if d>=0:
                    root_1 = round((-b + math.sqrt(d))/(2*a), 2)
                    root_2 = round((-b - math.sqrt(d))/(2*a), 2)
                    plt.plot([root_1, root_2],[0,0], 'go')
                    print(f'Roots: x = {int(root_1)} and x = {int(root_2)}')

                plt.show()
            quad()
            break

        elif user_choice == 5:
            def graph_functions():
                num = int(input("Number of functions to be graphed: "))
                for n in range (1, num + 1):
                    mb = input("Enter m and b, separated by a comma, to complete y = mx + b equation: ")
                    mb_split = mb.split(",")
                    m = float(mb_split[0])
                    b = float(mb_split[1])

                    xmin = -20
                    xmax = 20
                    ymin = -20
                    ymax = 20

                    points = 20*(xmax-xmin)
                    x = np.linspace(xmin,xmax,points)
                    
                    #print function(label) per plots
                    if b < 0:
                        minus_split = str(b).split("-")
                        neg_b = float(minus_split[1])
                        title = f'y = {int(m)}x - {int(neg_b)}'
                    else:
                        title = f'y = {int(m)}x + {int(b)}'

                    fig, ax = plt.subplots()
                    plt.axis([xmin,xmax,ymin,ymax]) 
                    plt.plot([xmin,xmax],[0,0],'k') 
                    plt.plot([0,0],[ymin,ymax], 'k') 

                    ax.set_title(title)
                    plt.plot(x, m*x + b)
                    plt.show()

            graph_functions()
            break

        elif user_choice == 6:
            def solve_2_eqs():

                x,y = symbols('x y')

                first = input("Enter the first equation. ('x' must be included; 'y' in at least one equation): 0 = ")
                second = input("Enter the second equation. ('x' must be included; 'y' in at least one equation): 0 = ")

                #print out equations
                print(f'First equation: {first} = 0')
                print(f'Second equation: {second} = 0\n')

                #get answer as coordinates
                solution = linsolve([first, second], (x, y))
                x_solution = solution.args[0][0]
                y_solution = solution.args[0][1]

                # output solution
                print(f'Solution: ({x_solution}, {y_solution})')
            solve_2_eqs()
            break

        elif user_choice == 7:
            def shading():
                mb = input("Enter m and b, separated by a comma, to complete y = mx + b equation: ")
                mb_split = mb.split(",")
                m = float(mb_split[0])
                b = float(mb_split[1])

                #create title for line plot
                if b < 0:
                    minus_split = str(b).split("-")
                    neg_b = float(minus_split[1])
                    title = f'y = {int(m)}x - {int(neg_b)}'
                else:
                    title = f'y = {int(m)}x + {int(b)}'

                xmin = -20
                xmax = 20
                ymin = -20
                ymax = 20

                points = 20*(xmax-xmin)
                x = np.linspace(xmin,xmax,points)

                fig, ax = plt.subplots()
                plt.axis([xmin,xmax,ymin,ymax]) 
                plt.plot([xmin,xmax],[0,0],'k') 
                plt.plot([0,0],[ymin,ymax], 'k') 
                plt.title(title)

                #plot line and shading
                y = m*x+b
                plt.plot(x, y,'b')
                plt.fill_between(x, y, ymax, facecolor='c')
            shading()
            break

        elif user_choice == 8:
            def sys_eqs():
                mb_1 = input("First equation: y = mx + b. Enter m and b, separated by a comma: ")
                mb_in1 = mb_1.split(",")
                m1 = float(mb_in1[0])
                b1 = float(mb_in1[1])

                mb_2 = input("Second equation: y = mx + b. Enter m and b, separated by a comma: ")
                mb_in2 = mb_2.split(",")
                m2 = float(mb_in2[0])
                b2 = float(mb_in2[1])

                #print out equations
                if b1 < 0 and b2 > 0:
                    minus_split = str(b1).split("-")
                    neg_b1 = float(minus_split[1])
                    print(f'First equation:   y = {int(m1)}x - {int(neg_b1)}')
                    print(f'Second equation:  y = {int(m2)}x + {int(b2)}')
                
                elif b1 > 0 and b2 < 0:
                    minus_split = str(b2).split("-")
                    neg_b2 = float(minus_split[1])
                    print(f'First equation:   y = {int(m1)}x + {int(b1)}')
                    print(f'Second equation:  y = {int(m2)}x - {int(neg_b2)}')

                elif b1 < 0 and b2 < 0:
                    minus_split = str(b1).split("-")
                    neg_b1 = float(minus_split[1])
                    minus_split = str(b2).split("-")
                    neg_b2 = float(minus_split[1])
                    print(f'First equation:   y = {int(m1)}x - {int(neg_b1)}')
                    print(f'Second equation:  y = {int(m2)}x - {int(neg_b2)}')

                else:
                    print(f'First equation:   y = {int(m1)}x + {int(b1)}')
                    print(f'Second equation:  y = {int(m2)}x + {int(b2)}')

                # solve the system of eqs
                x,y = symbols('x y')
                first = m1*x + b1 - y
                second = m2*x + b2 - y
                solution = linsolve([first, second], (x, y))
                x_solution = round(float(solution.args[0][0]),3)
                y_solution = round(float(solution.args[0][1]),3)

                #print out solution
                print(" ")
                print(f"Solution: ({x_solution}, {y_solution})")

                #plot graph
                xmin = int(x_solution) - 20
                xmax = int(x_solution) + 20
                ymin = int(y_solution) - 20
                ymax = int(y_solution) + 20
                points = 2*(xmax-xmin)

                #define x values once for the graph
                graph_x = np.linspace(xmin,xmax,points)

                # define y values for the graph
                y1 = m1*graph_x + b1
                y2 = m2*graph_x + b2

                fig, ax = plt.subplots()
                plt.axis([xmin,xmax,ymin,ymax]) 
                plt.plot([xmin,xmax],[0,0],'k') 
                plt.plot([0,0],[ymin,ymax], 'k') 

                # line 1
                plt.plot(graph_x, y1, 'b')

                # line 2
                plt.plot(graph_x, y2, 'r')

                # point
                plt.plot([x_solution],[y_solution],'go')
                plt.show()
            sys_eqs()
            break

        elif user_choice == 9:
            def zoom_graph():
                # input function to be plotted
                mb = input("Enter m and b, separated by a comma, to complete y = mx + b equation: ")
                mb_split = mb.split(",")
                m = float(mb_split[0])
                b = float(mb_split[1])

                #create title for line plot
                if b < 0:
                    minus_split = str(b).split("-")
                    neg_b = float(minus_split[1])
                    title = f'y = {int(m)}x - {int(neg_b)}'
                else:
                    title = f'y = {int(m)}x + {int(b)}'
                        
                def z(zoom):
                    
                    xmin = -zoom
                    xmax = zoom
                    ymin = -zoom
                    ymax = zoom
                    points = 10*xmax-xmin
                    x = np.linspace(xmin, xmax, points)

                    fig, ax = plt.subplots()
                    plt.axis([xmin,xmax,ymin,ymax]) 
                    plt.plot([xmin,xmax],[0,0],'black') 
                    plt.plot([0,0],[ymin,ymax], 'black') 
                    plt.plot(x, m*x + b)

                    ax.set_title(title)
                    ticks = int(round((xmax-xmin)/10))
                    ax.set_xticks(np.arange(xmin, xmax, ticks))
                    ax.set_yticks(np.arange(ymin, ymax, ticks))
                    ax.grid(True)
                    plt.show()

                #sliders
                interactive_plot = interactive(z, zoom=(1,100))
                display(interactive_plot)
            zoom_graph()
            break

        elif user_choice == 10:
            def quad():
                #enter quad values
                # \u00b2 prints 2 as an exponent
                a = float(input("For y = ax\u00b2 + bx + c, enter the 'a' value: "))
                b = float(input("For y = ax\u00b2 + bx + c, enter the 'b' value: "))
                c = float(input("For y = ax\u00b2 + bx + c, enter the 'c' value: "))

                #print out full quad eq
                if b < 0 and c < 0:
                    minus_split_b = str(b).split("-")
                    neg_b = float(minus_split_b[1])
                    minus_split_c = str(c).split("-")
                    neg_c = float(minus_split_c[1])
                    print(f"y = {int(a)}x\u00b2 - {int(neg_b)}x - {int(neg_c)}")
                elif b < 0:
                    minus_split = str(b).split("-")
                    neg_b = float(minus_split[1])
                    print(f"y = {int(a)}x\u00b2 - {int(neg_b)}x + {int(c)}")
                elif c < 0:
                    minus_split = str(c).split("-")
                    neg_c = float(minus_split[1])
                    print(f"y = {int(a)}x\u00b2 + {int(b)}x - {int(neg_c)}")
                else:
                    print(f"y = {int(a)}x\u00b2 + {int(b)}x + {int(c)}")

                # get vertex
                vx = -b/(2*a)
                vy = a*(vx**2) + b*vx + c
                print(f"Vertex: ({vx:,.2f}, {vy:,.2f})")

                # get roots
                d = b**2 - 4*a*c
                if d>=0:
                    root_1 = (-b + math.sqrt(d))/(2*a)
                    root_2 = (-b - math.sqrt(d))/(2*a)
                    print(f"Roots: x = {root_1:,.2f} and x = {root_2:,.2f}")
                else:
                    print("No real roots")
            quad()
            break

        else: 
            print("ENTRY ERROR. Number MUST be from 1 to 10\n")
             
graph_calc()
```
